```python
from sklearn.datasets import load_iris
import pandas as pd
iris=load_iris()
#iris.target
iris.feature_names
iris_data=pd.DataFrame(iris.data,columns=iris.feature_names)
#target=pd.DataFrame(iris.target)
#target.rename(columns={0:'Target'},inplace=True)
#iris_data.assign(Target=iris.target)

#Train an Agglomerative Clustering 

from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
clusters=model.fit_predict(iris.data)

from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
links = linkage(iris.data, 'complete')
plt.figure(figsize=(15,10))

dendrogram(links)

plt.show()
```

```python
plt.scatter(iris.data[clusters == 0, 0], iris.data[clusters == 0, 1], label = 'Type 1')
plt.scatter(iris.data[clusters == 1, 0], iris.data[clusters == 1, 1], label = 'Type 2')
plt.scatter(iris.data[clusters == 2, 0], iris.data[clusters == 2, 1], label = 'Type 3')
plt.legend()
plt.title('Clusters')
plt.show()
```

```python
from sklearn.datasets import load_iris

iris=load_iris()
X=iris.data
Y=iris.target
target_names=iris.target_names

print(X.shape)
target_names
```

```python
# We now perform Principal Component Analysis to reduce the features from four to two, for ease of visualization.
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np 
import pandas as pd


X = PCA(n_components=2).fit_transform(X)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
agg = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'complete') 
clusters = agg.fit_predict(X)
clusters
```

```python
plt.scatter(X[clusters == 0, 0], X[clusters == 0, 1], label = 'Type 1')
plt.scatter(X[clusters == 1, 0], X[clusters == 1, 1], label = 'Type 2')
plt.scatter(X[clusters == 2, 0], X[clusters == 2, 1], label = 'Type 3')
plt.legend()
plt.title('Clusters')
plt.show()
```

```python

#if we want 3 clusters then we have to make sure that the horizontal line cuts 3 vertical lines
from scipy.cluster.hierarchy import dendrogram, linkage

links = linkage(X, 'complete')
plt.figure(figsize=(15,10))

dendrogram(links)

plt.show()
```
