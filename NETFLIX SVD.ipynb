```python
df1=df[df['Ratings'].isna()==True].reset_index()

df1
```

```python

import numpy as np

movie_np = []
movie_id = 1

for i,j in zip(df1['index'][1:],df1['index'][:-1]):
    # numpy approach
    temp = np.full((1,i-j-1), movie_id)
    movie_np=np.append(movie_np, temp)
    movie_id += 1

# Account for last record and corresponding length
# numpy approach
last_record = np.full((1,len(df) - df1.iloc[-1, 0] - 1),movie_id)
movie_np = np.append(movie_np, last_record)

print(f'Movie numpy: {movie_np}')
print(f'Length: {len(movie_np)}')
```

```python
dataset = df[pd.notnull(df['Ratings'])]

dataset['Movie_Id'] = movie_np.astype('int')
dataset['CustId']=dataset['CustId'].astype('int')
#print('-Dataset examples-')
#dataset.head()
```

```python
dataset.shape
```

```python
#Most rated Movies
#f = ['count','mean']
dataset_movie_summary = dataset.groupby('Movie_Id')['Ratings'].count().reset_index()
dataset_movie_summary
movie_benchmark = round(dataset_movie_summary['Ratings'].quantile(0.7),0)

drop_movie_list = dataset_movie_summary[dataset_movie_summary['Ratings'] < movie_benchmark]

drop_movie_list

#print('Movie minimum times of review: {}'.format(movie_benchmark))
#dataset_movie_summary[dataset_movie_summary['CustId']<]
```
