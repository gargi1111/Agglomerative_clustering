```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get('https://www.cricbuzz.com/')
a=page.content
```

```python
soup=BeautifulSoup(a, 'html.parser')
```

```python
#Extract Title from all posts


titles=soup.find_all('h2',class_='big-crd-hdln')

titles_=[]
for title in titles:
    titles_.append(title.text)
    
titles_    
```

```python
#Extract Summary from all the posts

summary=soup.find_all('div',class_='cb-nws-intr')

summary_=[]
for sum_ in summary:
    summary_.append(sum_.text)
summary_    
```

```python
#Extract links 
links=[]
for link in soup.find_all('a'):
    links.append(link.get('href'))
links    
```

```python
data=[]
data.append(links)
data.append(summary_)
data.append(titles_)
```

```python
import pandas as pd

scrapped_data=pd.DataFrame(data)
scrapped_data.to_csv('scraped_data.csv',header=False,index=False)

```
